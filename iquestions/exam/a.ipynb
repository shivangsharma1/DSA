```python
def binning_func(data_row):
    if data_row["tenure"]>=0 and data_row["tenure"]<24:
        return "group1"
    if data_row["tenure"]>=24 and data_row["tenure"]<48:
        return "group2"
    if data_row["tenure"]>=48 and data_row["tenure"]<60:
        return "group3"
    if data_row["tenure"]>=60:
        return "group4"
```

```python
charges_data["tenureBinned"] = charges_data.apply(lambda x: binning_func(x), axis = 1)
```
